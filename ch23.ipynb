```python
from skimage import io
from skimage.morphology import white_tophat, disk
from skimage.transform import rescale
import numpy as np
import matplotlib.pyplot as plt
import time

image = io.imread('rbexample.png')

image_downsampled = rescale(image, 0.5, anti_aliasing=True)

ball_size = 10
selem = disk(ball_size)

start_time = time.time()

background_subtracted_downsampled = white_tophat(image_downsampled, selem)

background_subtracted = rescale(background_subtracted_downsampled, 2.0, anti_aliasing=True)

without_background = image - background_subtracted

end_time = time.time()
time_taken = end_time - start_time

fig, axes = plt.subplots(1, 3, figsize=(18, 6))
axes[0].imshow(image, cmap='gray')
axes[0].set_title('Input Image')
axes[0].axis('off')

axes[1].imshow(background_subtracted, cmap='gray')
axes[1].set_title('Subtracted Background')
axes[1].axis('off')

axes[2].imshow(without_background, cmap='gray')
axes[2].set_title('Without Background')
axes[2].axis('off')

plt.show()

time_taken
```
